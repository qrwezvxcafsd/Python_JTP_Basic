```python
# tuples can also be added (덧셈) or multiplied just like lists, but can't be
# sorted.

tuple1 = (0, 1, 2, 3)
tuple2 = ('python', 'java')

print(tuple1 + tuple2)

t1 = (1, 2, 3)
t2 = t1 * 3
print(t2)

t1 = (1, 2, 3)
print(t1[1])
```

```python
# 02.05 dictionary
# dictionary - each key has its own value. (이름: 유홍현, 나이: 22, 주소: 분당구 정자동)
# dic = {key : value}
```

```python
dic = {'name' : '유홍현', 'age' : 22, 'address' : '분당구 정자동'}
print(dic)
print(type(dic))
```

```python
# adding components of dictionary

a = {1 : 'a'}
a[2] = 'b'
a[4] = [1, 2, 3]
print(a)
```

```python
# deleting components of dictionary

del a[1]
print(a)
```

```python
# getting Key using Value of dictionary

dic = {'name' : '유홍현', 'age' : 22, 'address' : '분당구 정자동'}
print(dic.keys())
print(dic.values())

print(list(dic.keys()))
print(list(dic.values()))
```

```python
# 교집합, 차집합, 합집합 등 구할 수 있음
```

```python
# 교집합 (intersection)

s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1 & s2)
```

```python
# 합집합 (union)

print(s1 | s2)
```

```python
# 차집합 (difference)

print(s1 - s2)
```

```python
# add

s1.add(20)
print(s1)
```

```python
# update

s1.update([21,22,23])
print(s1)
```

```python
# remove

s1.remove(1)
print(s1)
```
