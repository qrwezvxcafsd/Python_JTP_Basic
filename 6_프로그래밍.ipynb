```python
# 06.01 How to build a programme

# Keep thinking inputs and outputs to bulid a programme
# Ex. Creating a multipling programme

# 1. What should the name of a funcion be? -> gugu
# 2. What is the input? -> 2
# 3. What are the outputs? -> 2, 4, 6, ..., 18
# 4. In which form should the outputs be saved?
# -> since they are countinued, 'List'
```

```python
# primitive way

def gugu(n):
    result = []
    result.append(n*1)
    result.append(n*2)
    result.append(n*3)
    result.append(n*4)
    result.append(n*5)
    result.append(n*6)
    result.append(n*7)
    result.append(n*8)
    result.append(n*9)
    return result

print(gugu(2))
print(gugu(7))
```

```python
# sophiscated way

def gugu(n):
    result = []
    for i in range(1, 10):
        result.append(n*i)
    return result

print(gugu(2))
print(gugu(7))
```

```python
# 06.02 Find the sum of all multiplies of 3 and 5

# Q. Find sum of all multipilies of 3 and 5 (N<1000)
```

```python
# Hong's solution

for i in range (1, 1000):
    a = 0
    if i % 3 ==0 or i % 5 == 0:
        a = a + i
    elif i % 15 == 0:
        a = a - i

print(a)
```

```python
# Hong's solution fixed

a = 0     # a outside the loop

for i in range (1, 1000):
    if i % 3 == 0 or i % 5 == 0:
        a = a + i

print(a)
```

```python
# 06.03 Paging a board
# 20 articles per page
# Want to make a programme that gives the number of total pages
# when the number of articles is given
```

```python
# Hong's solution

a = int(input('How many articles? '))

b = a // 20

if a % 20 == 0:
    if a == 0:
        b = 1
    else:
        b = b
elif a % 20 != 0:
    b = b + 1

print('There are %d pages' %b)
```

```python
# Youtube solution

def get_total_page(m, n):
    if m % n == 0:
        return m // n
    else:
        return m // n + 1

print(get_total_page(20, 20))
print(get_total_page(21, 20))
print(get_total_page(0, 20))
print(get_total_page(25637, 20))
```
