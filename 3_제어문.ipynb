```python
# How can I get back to the start pint of 'while'?

a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0:
        continue
    print(a)
print('The programme is over')
```

```python
# 03.03 'for' sentences
```

```python
TL = {'one', 'two', 'three'}
for x in TL:
    print(x)
```

```python
# list that has tuples as components

a = [(1, 2), (3, 4), (5, 6)]
for (x,y) in a:
    print(x+y)
```

```python
# practical usage of 'for' loop

grades = [90, 25, 67, 45, 80] # grades of students
number = 0 # numbers for each student
for grade in grades:
    number = number + 1
    if grade >= 60:
        print('student %d is passed.' % number)
    else:
        print('student %d is failed.' % number)
```

```python
# continue in 'for' loops

grades = [90, 25, 67, 45, 80] # grades of students
number = 0 # numbers for each student
for grade in grades:
    number = number + 1
    if grade < 60:
        continue
    print('student %d is passed.' % number)
```

```python
# using Range in 'for' loops

add = 0
for i in range(1, 11):
    add = add + i
    print(add)
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ") # end=" " means empty space
    print('')
```
