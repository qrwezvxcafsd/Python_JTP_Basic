```python
a=123
print(type(a))

b=1.2
print(type(b))

c=1.2e3
print(type(c))
print(c)

d=0o10
print(type(d))
print(d)

# a의 타입이 int이다. (integer, 정수)
# b의 타입이 float이다. (실수)
# c는 컴퓨터식 지수 표현 방식(e랑 E 둘 다 사용 가능), 실수형으로 처리
# d는 8진수 표현법
```

```python
# 사칙연산 가능
# 덧셈(+) 곱셈(*) 나눗셈(/) 몫(//) 지수(**) 나머지(%)

a=3
b=4

print(a+b)
print(a*b)
print(a/b)
print(a//b)
print(a**b)
print(a%b)
```

```python
# 따옴표를 3개 연속으로 쓰면, 여러줄로 써도 하나의 문자열로 인식
# \(백슬레시)는 특수기호 넣을 때 쓰는 것
# \n은 줄바꿈, \t는 tab 역할, \\는 \를 글자로 쓰고싶을 때
```

```python
#문자열 연산
```

```python
#문자열 더해서 연산하기

head = "Python"
tail = " is fun!"
print(head + tail)
```

```python
# 문자열 곱하기

a = "python"
print(a * 2)
```

```python
# 문자열 길이 구하기 (len function) length

a = "Life is too short"
print(len(a))
```

```python
# 문자열 인덱싱
# 인덱싱은 []

a = "Life is too short, you need Python"
print(a[3])
print(a[0])
print(a[-1])
```

```python
# 문자열 슬라이싱
# [이상:미만:간격]

a = "Life is too short, you need Python"
print(a[0:4])
print(a[19:])
print(a[:17])
print(a[19:-7])
print(a[::-2])
```

```python
# 문자열 포매팅
# %s는 다 넣을 수 있음(숫자나 뭐나 다)

a = "I eat %d apples." %3 # decimal(십진수)
print(a)

b = "I eat %s apples." %"five" # string
print(b)

number = 3
day = "three"
c = "I ate %d apples. so I was sick for %s days." %(number, day)
print(c)

d = "I ate %s apples. so I was sick for %d days." %(day, number)
print(d)
```

```python
# f 포매팅 (제일 최근)

name = '유홍현'
age = 22
a = f'나의 이름은 {name}입니다. 나이는 {age}살입니다.'
b = f"나는 내년이면 {age+1}살이 된다."

print(b)
print(a)
```

```python
# 문자열 개수 세기

a = "Hobby"
print(a.count('b'))
```

```python
# 위치 알려주기

a = "Hence"
print(a.find('e'))
```

```python
# 문자 삽입

a = ", ".join('abcd')
print(a)
```

```python
# 공백 없애기 (strip)

a = ' Hi '

print(a.lstrip())
print(a.rstrip())
print(a.strip())
```

```python
# 02.3 리스트 - 변수가 상자라면 리스트는 서랍장 (1번서랍 2번서랍 등등에 각각 값 지정 가능)
```

```python
odd = [1, 3, 5,7,9]
print(type(odd))

e = [1, 2, 'lie']
print(e)
print(type(e))
```

```python
a = [1, 'hello', ['a', 2]]

print(a[1][2])
print(a[2][1])
print(a[0:2])
```

```python
# 리스트 연산

a = [1, 2, 3]
b = [4, 5, 6]

print(a+b)
print(a*3)
```

```python
# 리스트 길이 구하기 (len) 이용

a = [1, 2, 3]

print(len(a))
```

```python
# 리스트 값 바꿔치기

a = [1, 2, 3]
a[1] = '홍현'

print(a)
```

```python
# 리스트 값 없애기 (del)

a = [1, 2, 3]
del a[1]

print(a)

b = [1, 2, 3, 4, 5]
del b[2:]

print(b)
```

```python
# 리스트 값 추가하기

a = [1,2,3]
a.append(4)

print(a)
```

```python
# 리스트 정렬 (sort)

a = [1, 3,2]
b = ['s', 'b', 'd']

a.sort()
b.sort()

print(a)
print(b)
```
